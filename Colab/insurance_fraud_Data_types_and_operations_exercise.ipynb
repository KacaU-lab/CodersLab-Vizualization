Zjisti:

kter√© sloupce dataset obsahuje,

kolik hodnot null (NaN) je v jednotliv√Ωch sloupc√≠ch,

jak√© maj√≠ datov√© typy.

Pomoc√≠ astype() zmƒõ≈à datov√© typy sloupc≈Ø DayOfWeek, Make a AccidentArea z object na category.

Ve sloupci DayOfWeek zmƒõ≈à v≈°echna p√≠smena na velk√°.

Ve sloupci Make zmƒõ≈à v≈°echna p√≠smena na mal√° a vyfiltruj pouze ty ≈ô√°dky, kde text obsahuje "hon" (nap≈ô. "honda").

V√Ωsledek ulo≈æ do promƒõnn√© frauds_df_honda.

Zobraz frauds_df_honda.

P≈ôidej nov√Ω sloupec PolicyRepNumber, kter√Ω bude souƒçtem sloupc≈Ø PolicyNumber a RepNumber.

```python
# 1Ô∏è‚É£ Import pot≈ôebn√Ωch knihoven
import pandas as pd

# 2Ô∏è‚É£ Naƒçten√≠ datov√©ho souboru (nap≈ô. frauds.csv)
# Colab vyzve k nahr√°n√≠ souboru
from google.colab import files
uploaded = files.upload()

# P≈ôedpokl√°dejme, ≈æe soubor se jmenuje "frauds.csv"
frauds_df = pd.read_csv("frauds.csv")

# 3Ô∏è‚É£ Z√°kladn√≠ p≈ôehled o datech
print("üîπ Sloupce v datasetu:")
print(frauds_df.columns)

print("\nüîπ Poƒçet chybƒõj√≠c√≠ch hodnot (NaN):")
print(frauds_df.isnull().sum())

print("\nüîπ Datov√© typy:")
print(frauds_df.dtypes)

# 4Ô∏è‚É£ Zmƒõna datov√Ωch typ≈Ø na 'category'
cols_to_convert = ['DayOfWeek', 'Make', 'AccidentArea']
for col in cols_to_convert:
    frauds_df[col] = frauds_df[col].astype('category')

print("\n‚úÖ Typy po konverzi:")
print(frauds_df.dtypes[cols_to_convert])

# 5Ô∏è‚É£ Sloupec DayOfWeek ‚Äì v≈°echna p√≠smena VELK√Å
frauds_df['DayOfWeek'] = frauds_df['DayOfWeek'].str.upper()

# 6Ô∏è‚É£ Sloupec Make ‚Äì v≈°echna p√≠smena mal√° + filtr na 'hon'
frauds_df['Make'] = frauds_df['Make'].str.lower()

frauds_df_honda = frauds_df[frauds_df['Make'].str.contains('hon', na=False)]

# 7Ô∏è‚É£ V√Ωpis v√Ωsledn√©ho v√Ωbƒõru
print("\nüöó DataFrame pro znaƒçku 'Honda' (obsahuje 'hon'):")
print(frauds_df_honda.head())

# 8Ô∏è‚É£ Nov√Ω sloupec: PolicyRepNumber = PolicyNumber + RepNumber
frauds_df_honda['PolicyRepNumber'] = frauds_df_honda['PolicyNumber'] + frauds_df_honda['RepNumber']

print("\nüÜï P≈ôidan√Ω sloupec 'PolicyRepNumber':")
print(frauds_df_honda[['PolicyNumber', 'RepNumber', 'PolicyRepNumber']].head())
```

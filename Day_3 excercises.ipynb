```python
# 1) Naƒçten√≠ iris datov√© sady z plotly.express
df = px.data.iris()

print("Prvn√≠ch 5 ≈ô√°dk≈Ø iris datasetu:")
print(df.head())

# 2) Vytvo≈ôen√≠ bodov√©ho grafu:
#    - osa x: sepal_length
#    - osa y: sepal_width
#    - barva: species (ka≈æd√Ω druh jinou barvou)
fig = px.scatter(
    df,
    x="sepal_length",
    y="sepal_width",
    color="species",
    title="Iris: vztah mezi d√©lkou a ≈°√≠≈ôkou kali≈°n√≠ho l√≠stku"
)
```

```python
from google.colab import drive
drive.mount('/content/drive')

import os

base_path = "/content/drive/MyDrive/Colab Notebooks/Vis"
print(f">> Prohled√°v√°m slo≈æku: {base_path}")

csv_jpg_files = []
for root, dirs, files in os.walk(base_path):
    for file in files:
        if file.lower().endswith((".csv", ".jpg", ".jpeg")):
            csv_jpg_files.append(os.path.join(root, file))

print("\nüìÇ Nalezen√© CSV a JPG soubory:")
for path in csv_jpg_files:
    print(" -", path)

print("\n‚úÖ √övodn√≠ bu≈àka probƒõhla v po≈ô√°dku ‚Äì Drive p≈ôipojen a soubory vyps√°ny.")
```

```python
# Instalace a importy
print(">> Instaluji / importuji knihovny pro Dash...")

!pip install dash==2.17.0 --quiet

import plotly.express as px
from dash import Dash, dcc, html, Input, Output

import threading
from google.colab import output

print("‚úÖ Dash a pot≈ôebn√© moduly importov√°ny.")

# 1) Naƒçten√≠ iris datasetu
df = px.data.iris()
print(">> Naƒçtena iris data, prvn√≠ch 5 ≈ô√°dk≈Ø:")
print(df.head())

# 2) Jedineƒçn√Ω seznam species
species_list = sorted(df["species"].unique())
print("\n>> Jedineƒçn√© hodnoty species:")
print(species_list)

# 3) Vytvo≈ôen√≠ v√Ωchoz√≠ho grafu (pro prvn√≠ species)
default_species = species_list[0]
print(f"\n>> V√Ωchoz√≠ species pro zobrazen√≠: {default_species}")

def create_figure_for_species(species_name: str):
    """Pomocn√° funkce ‚Äì vytvo≈ô√≠ scatter jen pro dan√Ω species."""
    filtered = df[df["species"] == species_name]
    fig = px.scatter(
        filtered,
        x="sepal_length",
        y="sepal_width",
        color="species",  # po≈ô√°d barevnƒõ rozli≈°√≠ druh (bude jedna barva)
        title=f"Iris ‚Äì sepal_length vs. sepal_width ({species_name})"
    )
    return fig

fig_default = create_figure_for_species(default_species)
print("‚úÖ V√Ωchoz√≠ figure pro Dash aplikaci p≈ôipraven.")

# 4) Vytvo≈ôen√≠ Dash aplikace
app = Dash(__name__)
print(">> Objekt Dash aplikace vytvo≈ôen.")

# 5) Layout aplikace s H1, H3 a Dropdownem + Graph
app.layout = html.Div(children=[
    html.H1("Hello dash"),
    html.H3("Choose species:"),

    dcc.Dropdown(
        id="species-dropdown",
        options=[{"label": s.title(), "value": s} for s in species_list],
        value=default_species,  # v√Ωchoz√≠ hodnota
        clearable=False        # nelze ‚Äûvymazat‚Äú v√Ωbƒõr
    ),

    dcc.Graph(
        id="iris-graph",
        figure=fig_default
    )
])

print("‚úÖ Layout aplikace byl nastaven (H1, H3, Dropdown, Graph).")

# 6) Callback ‚Äì reaguje na zmƒõnu hodnoty v Dropdownu
@app.callback(
    Output("iris-graph", "figure"),
    Input("species-dropdown", "value")
)
def update_iris_graph(selected_species):
    print(f">> Callback: vybr√°n species = {selected_species}")
    fig = create_figure_for_species(selected_species)
    return fig

print("‚úÖ Callback pro aktualizaci grafu podle species byl zaregistrov√°n.")
print("‚úÖ Bu≈àka s definic√≠ Dash aplikace probƒõhla √∫spƒõ≈°nƒõ.")
```

```python
print(">> P≈ôipravuji spu≈°tƒõn√≠ Dash serveru ve vl√°knƒõ...")

port = 8050  # m≈Ø≈æe≈° zmƒõnit, kdyby nƒõco kolidovalo

def run_dash():
    print(">> [vl√°kno] Startuji Dash server...")
    app.run_server(
        host="0.0.0.0",
        port=port,
        debug=False,
        use_reloader=False
    )

thread = threading.Thread(target=run_dash, daemon=True)
thread.start()

print(f"‚úÖ Dash bƒõ≈æ√≠ na portu {port}")
print(">> Zobrazuji aplikaci v iframe v Colabu...")

# doporuƒçen√Ω zp≈Øsob ‚Äì iframe
output.serve_kernel_port_as_iframe(port)

print("‚úÖ Dash aplikace byla √∫spƒõ≈°nƒõ zobrazena v notebooku.")
```

```python
print(">> Importuji knihovny a naƒç√≠t√°m data...")

import pandas as pd
import plotly.express as px

df = px.data.tips()
print(">> Data naƒçtena.")

print("\nüìå Struktura datasetu:")
print(df.info())

print("\nüìå N√°hled:")
print(df.head())

# 1. P≈ôid√°n√≠ mapovan√©ho dne
df["day_num"] = df["day"].map({"Thur": 1, "Fri": 2, "Sat": 3, "Sun": 4})

# 2. Agregace
revenue_per_day = df.groupby("day", as_index=False)["total_bill"].sum()
avg_tip_per_day = df.groupby("day", as_index=False)["tip"].mean()

print("\nüìå Denn√≠ tr≈æby:")
print(revenue_per_day)

print("\nüìå Pr≈Ømƒõrn√© spropitn√©:")
print(avg_tip_per_day)

# Plotly grafy (vyu≈æije se i jako fallback)
fig_revenue = px.bar(revenue_per_day, x="day", y="total_bill", title="Tr≈æby podle dne")
fig_tip_day = px.bar(avg_tip_per_day, x="day", y="tip", title="Pr≈Ømƒõrn√© spropitn√© podle dne")

print("‚úÖ Data p≈ôipravena a grafy vytvo≈ôeny.")
```

```python
import pandas as pd
import plotly.express as px
from dash import Dash, dash_table, html
from google.colab import output
import threading

print(">> Naƒç√≠t√°m dataset tips...")
df = px.data.tips()

# 1) Pivot table (index = day, columns = sex, values = total_bill)
print(">> Vytv√°≈ô√≠m kontingenƒçn√≠ tabulku...")
pivot = pd.pivot_table(
    df,
    index="day",
    columns="sex",
    values="total_bill",
    aggfunc="sum"
)

print("P≈Øvodn√≠ pivot:")
print(pivot)

# 2) Zmƒõna po≈ôad√≠ ≈ô√°dk≈Ø pomoc√≠ reindex
print(">> Mƒõn√≠m po≈ôad√≠ ≈ô√°dk≈Ø...")
pivot = pivot.reindex(["Thur", "Fri", "Sat", "Sun"])

# 3) Zmƒõna typu sloupc≈Ø na string
pivot.index = pivot.index.astype(str)
pivot.columns = pivot.columns.astype(str)

print("Upraven√Ω pivot:")
print(pivot)
```

```python
app = Dash(__name__)

# --- Podm√≠nƒõn√© form√°tov√°n√≠ ---
style_conditional = [
    {
        'if': {'filter_query': '{Female} >= 100', 'column_id': 'Female'},
        'backgroundColor': 'green',
        'color': 'white'
    },
    {
        'if': {'filter_query': '{Male} >= 100', 'column_id': 'Male'},
        'backgroundColor': 'royalblue',
        'color': 'white'
    }
]

# --- Layout ---
app.layout = html.Div([
    html.H3("Kontingenƒçn√≠ tabulka ‚Äì √∫traty podle dne a pohlav√≠"),

    dash_table.DataTable(
        id='pivot-table',
        columns=[{"name": c, "id": c} for c in pivot.columns],
        data=pivot.reset_index().to_dict("records"),
        style_table={'width': '500px'},
        style_data_conditional=style_conditional,
        export_format="csv",
        export_headers="display"
    )
])
```

```python
def run_app():
    app.run_server(port=8051, debug=False, use_reloader=False)

thread = threading.Thread(target=run_app)
thread.start()

print("üìå Dash bƒõ≈æ√≠ na http://127.0.0.1:8051")

output.serve_kernel_port_as_iframe(8051, height=550)
```

```python
import base64
import io
import pandas as pd

from dash import Dash, dcc, html, dash_table, Input, Output
from jupyter_dash import JupyterDash

# ====== Styl upload komponenty (ze zad√°n√≠) ======
upload_style = {
    'width': '100%',
    'height': '60px',
    'lineHeight': '60px',
    'borderWidth': '1px',
    'borderStyle': 'dashed',
    'borderRadius': '5px',
    'textAlign': 'center',
    'margin': '10px'
}

# ====== Vytvo≈ôen√≠ aplikace ======
app = JupyterDash(__name__)

app.layout = html.Div([
    html.H2("Naƒç√≠t√°n√≠ CSV souboru do tabulky"),

    dcc.Upload(
        id="upload-data",
        children=html.Div(["P≈ôet√°hnƒõte CSV sem nebo kliknƒõte"]),
        style=upload_style,
        multiple=False
    ),

    dash_table.DataTable(
        id="output-table",
        page_size=10,
        style_table={'overflowX': 'auto'},
    )
])


# ====== CALLBACK ‚Äî zpracov√°n√≠ CSV ======
@app.callback(
    Output("output-table", "data"),
    Output("output-table", "columns"),
    Input("upload-data", "contents"),
    Input("upload-data", "filename")
)
def update_table(contents, filename):
    if contents is None:
        return [], []

    # dek√≥dov√°n√≠ CSV
    content_type, content_string = contents.split(',')
    decoded = base64.b64decode(content_string)

    df = pd.read_csv(io.BytesIO(decoded))

    return df.to_dict('records'), [{"name": c, "id": c} for c in df.columns]


# ====== Spu≈°tƒõn√≠ v Colabu ======
app.run_server(mode="inline", height=600)
```

Draw a simple graph for which the list of edges is written as a list of tuples. The node labels are letters.

Use the data below:

edge_list = [('B', 'A'), ('C', 'B'), ('B', 'E'), ('A', 'C'), ('C', 'D'), ('D', 'E')]

First, draw the graph with a circular_layout and then with a random_layout. Use the variable:

pos = nx.spring_layout(G)



```python
import networkx as nx
import matplotlib.pyplot as plt

# Edge list
edge_list = [('B', 'A'), ('C', 'B'), ('B', 'E'),
             ('A', 'C'), ('C', 'D'), ('D', 'E')]

# Create an undirected graph
G = nx.Graph()
G.add_edges_from(edge_list)

# 1. Draw with circular_layout
plt.figure(figsize=(4, 4))
pos_circular = nx.circular_layout(G)
nx.draw_networkx(G, pos=pos_circular, with_labels=True)
plt.title("Graph with circular_layout")
plt.axis("off")
plt.show()

# 2. Draw with random_layout
plt.figure(figsize=(4, 4))
pos_random = nx.random_layout(G)
nx.draw_networkx(G, pos=pos_random, with_labels=True)
plt.title("Graph with random_layout")
plt.axis("off")
plt.show()

# Optional: spring_layout stored in variable pos
pos = nx.spring_layout(G)
# Now you can do for example:
plt.figure(figsize=(4, 4))
nx.draw_networkx(G, pos=pos, with_labels=True)
plt.title("Graph with spring_layout (pos)")
plt.axis("off")
plt.show()
```

```python
@app.callback(
    Output('graph', 'figure'),
    [Input("template-dropdown", "value")]
)
def update_figure(template):
    chart = px.scatter(df, x="total_bill", y="tip", template=template,
        color="sex", size="size", render_mode="webgl", title="Total bill vs tip amounts",
        labels={"tip": "tip amount", "total_bill": "total bill", "day": "transaction day"}
    )
    return chart
```

```python
# 2. Imports from dash and plotly.express
from dash import Dash, dcc, html
import plotly.express as px

# 3. Imports needed to display a Dash app in Colab
import threading
from google.colab import output

# 4. Load the iris dataset from plotly.express
df = px.data.iris()

# 5. Create a scatter plot
#    Visualize sepal length vs sepal width with different colors for each species
fig = px.scatter(
    df,
    x="sepal_length",
    y="sepal_width",
    color="species",
    title="Sepal length vs sepal width by species"
)

# 6. Create a Dash app object
app = Dash(__name__)

# 7. Define the app layout
app.layout = html.Div(children=[
    html.H1("Hello dash"),
    html.H3("Dash: A web application framework for Python."),
    dcc.Graph(
        id="iris-scatter-graph",  # any id you choose
        figure=fig
    )
])

# 8. Run Dash in a background thread and show it in Colab using an iframe
def run_dash():
    app.run_server(
        host="0.0.0.0",
        port=8050,
        debug=False,
        use_reloader=False
    )

thread = threading.Thread(target=app.run)
thread.start()
output.serve_kernel_port_as_iframe(port=8050, height='600')
```

The goal of this exercise is to create a simple Dash application. The data we will use in the application is the iris dataset available in the plotly.express package. Download it with: df=px.data.iris().

The application should contain 2 charts:

scatter plot - showing the relationship between the width and length of the iris flower sepal (sepal_width and sepal_length)
line chart - showing the values of the sepal length on the y-axis.
To create this application, follow the steps:

In the application, create two blocks using the html.Div tag. The blocks are to be displayed inline. Set the width of the blocks to 40% and the height to 500 pixels.
Put each of the two charts in its own block.

```python
# Imports
from dash import Dash, dcc, html
import plotly.express as px
import threading
from google.colab import output

# Data
df = px.data.iris()

# Charts
scatter_fig = px.scatter(
    df,
    x="sepal_width",
    y="sepal_length",
    color="species",
    title="Sepal width vs sepal length"
)

line_fig = px.line(
    df,
    y="sepal_length",
    title="Sepal length line chart"
)

# Dash app
app = Dash(__name__)

app.layout = html.Div([
    html.H1("Hello Dash"),
    html.Div("Dash: A web application framework for Python."),

    html.Div([
        dcc.Graph(
            id="scatter-plot",
            figure=scatter_fig
        )
    ],
    style={
        "display": "inline-block",
        "width": "40%",
        "height": "500px",
        "vertical-align": "top"
    }),

    html.Div([
        dcc.Graph(
            id="line-chart",
            figure=line_fig
        )
    ],
    style={
        "display": "inline-block",
        "width": "40%",
        "height": "500px",
        "vertical-align": "top",
        "margin-left": "3%"
    }),

])

# Run server through threading
def run_dash():
    app.run_server(
        host="0.0.0.0",
        port=8050,
        debug=False,
        use_reloader=False
    )

thread = threading.Thread(target=app.run)
thread.start()
output.serve_kernel_port_as_iframe(port=8050, height='600')
```

Nainstalujte verzi 4.14.3 knihovny plotly.

Naimportujte modul plotly.express a z plotly také graph_objects.

Načtěte data:

s výsledky voleb – pomocí px.data.election(),

s polygony popisujícími jednotlivé okresy – pomocí px.data.election_geojson().

Přiřaďte nějaké proměnné svůj přístupový token k Mapboxu.

Vytvořte vizualizaci, která:

zobrazuje počet hlasů pro kandidáta Coderre v každém okrese,

používá barevnou paletu Viridis,

barvu odlišuje podle hodnot proměnné Coderre,

nastaví střed mapy na souřadnice zeměpisné šířky 45.5517 a délky -73.7073,

má počáteční zoom 8.5,

používá dark styl mapy.

Tip: v definici go.Choroplethmapbox je potřeba zadat parametr featurekeyid (správně featureidkey), který odpovídá názvu okresu v GeoJSON datech.

```python
# 1. Instalace / importy
# (v novém prostředí případně: !pip install plotly==5.24.0)
import plotly.express as px
import plotly.graph_objects as go

# 2. Načtení vestavěných dat Plotly
election = px.data.election()              # tabulka s výsledky
geojson = px.data.election_geojson()       # GeoJSON s okrsky

# Krátká kontrola struktur
print(election.head())
print("Počet okresů v tabulce:", len(election))
print("Počet polygonů v GeoJSON:", len(geojson["features"]))

# 3. Mapbox token – DOPLŇ SI VLASTNÍ
MAPBOX_TOKEN = "YOUR_MAPBOX_TOKEN_HERE"

# 4. Vytvoření choropleth mapy v graph_objects
fig = go.Figure(
    go.Choroplethmapbox(
        geojson=geojson,
        locations=election["district"],      # propojovací klíč z tabulky
        z=election["Coderre"],              # Coderre = počet hlasů (numerická hodnota)
        featureidkey="properties.district", # propojovací klíč z GeoJSONu
        colorscale="Viridis",               # požadovaná paleta
        marker_line_width=0.5,
        marker_line_color="white",
        colorbar_title="Hlasy pro Coderre"
    )
)

# 5. Nastavení rozložení a stylu mapy
fig.update_layout(
    mapbox=dict(
        accesstoken=MAPBOX_TOKEN,
        style="dark",                        # dark styl
        center=dict(lat=45.5517, lon=-73.7073),
        zoom=8.5
    ),
    margin=dict(l=0, r=0, t=0, b=0),
    title="Počet hlasů pro kandidáta Coderre podle okresů (Montreal)"
)

fig.show()
```

Načte knihovnu plotly.express (kvůli datům) a plotly.graph_objects (kvůli Choroplethmapbox).

Načte data election a election_geojson() – žádné soubory neřešíš.

Připraví proměnnou MAPBOX_TOKEN, kam vložíš svůj token z Mapboxu.

go.Choroplethmapbox:

geojson – tvar okresů,

locations – jména okresů z tabulky,

z – počet hlasů pro Coderre,

featureidkey="properties.district" – říká, že v GeoJSONu je klíč v feature["properties"]["district"].

colorscale="Viridis" – přesně jak chce zadání.

mapbox.style="dark", center=(45.5517, -73.7073), zoom=8.5 – nastavení vzhledu a zoomu mapy.

Jak to interpretovat

Každý polygon (okres) je vybarven podle počtu hlasů pro Coderre.

Tmavší / světlejší barvy (dle palety Viridis) umožní rychle vidět, kde měl Coderre vyšší / nižší podporu.

Protože jde o choropleth mapu, plocha polygonu nereprezentuje počet hlasů – jen barva (intenzita).

Střed mapy a zoom jsou nastavené tak, aby se dobře zobrazil Montreal v základním náhledu.

```python
# (Colab) instalace a nastavení widgetů
!pip install -q plotly ipywidgets

from google.colab import output
output.enable_custom_widget_manager()

import plotly.express as px
import plotly.graph_objects as go
import ipywidgets as widgets
from IPython.display import display
```

```python
# --- 1) Načtení demo dat z Plotly ---
election = px.data.election()            # tabulka s výsledky voleb
geojson = px.data.election_geojson()     # GeoJSON s polygony okresů

# pro jistotu se podívej na první řádky (není povinné)
display(election.head())

# --- 2) Widget pro zadání Mapbox tokenu ---
token_input = widgets.Password(
    description='Mapbox token:',
    placeholder='vložit token sem',
    layout=widgets.Layout(width='60%')
)

btn_draw = widgets.Button(
    description='Vykreslit mapu',
    button_style='primary'
)

out = widgets.Output()

def draw_map(_):
    out.clear_output()
    token = token_input.value.strip()
    if not token:
        with out:
            print("⚠️ Zadej prosím Mapbox access token.")
        return

    # --- 3) Vytvoření Choroplethmapbox podle zadání ---
    fig = go.Figure(
        go.Choroplethmapbox(
            geojson=geojson,
            locations=election["district"],      # propojení s GeoJSON
            z=election["Coderre"],               # počet hlasů pro Coderre
            colorscale="Viridis",                # barevná paleta
            featureidkey="properties.district",  # název okresu v GeoJSON
            marker_opacity=0.7,
            marker_line_width=0
        )
    )

    fig.update_layout(
        mapbox_style="dark",                     # dark map style
        mapbox_accesstoken=token,               # tady se použije zadaný API klíč
        mapbox_center={"lat": 45.5517, "lon": -73.7073},
        mapbox_zoom=8.5,
        margin={"r":0, "t":0, "l":0, "b":0}
    )

    with out:
        fig.show()          # v Colabu opravdu MUSÍ být fig.show()

btn_draw.on_click(draw_map)

display(token_input, btn_draw, out)
```

Nainstalujte knihovnu folium.

Importujte knihovnu folium a konstruktor Figure z knihovny branca.

Vytvořte objekt Figure s rozměry: šířka 650 px a výška 400 px. Přiřaďte obrázek do proměnné fig.

Vytvořte objekt mapy se středem na souřadnicích 52.232418, 21.007151 (Varšava). Nastavte počáteční přiblížení na 12.

Připojte mapu k dříve definované proměnné fig.

Přidejte na mapu dvě další vrstvy dlaždic: cartodbdark_matter a cartodbpositron. Přidejte panel, který umožní přepínat mezi vrstvami.

Přidejte na mapu libovolné tři body ve Varšavě. Do konstruktoru značek předejte informace o souřadnicích a do parametru popup popis označeného objektu. Dále použijte parametr tooltip pro text, který se zobrazí po najetí myší.

Zobrazte mapu.

```python
# --- 1. Instalace knihoven (v Colabu většinou stačí jednou) ---
# Pokud se ti folium / branca nenačítají, odkomentuj:
# !pip install folium branca

# --- 2. Importy podle zadání ---
import folium
from branca.element import Figure

# --- 3. Figure objekt 650 x 400 px ---
fig = Figure(width=650, height=400)

# --- 4. Střed mapy: souřadnice Varšavy + zoom 12 ---
warsaw_center = [52.232418, 21.007151]

# Tiles=None, protože vrstvy chceme přidat ručně v dalším kroku
m = folium.Map(location=warsaw_center, zoom_start=12, tiles=None)

# --- 5. Připojení mapy do Figure ---
fig.add_child(m)

# --- 6. Dlaždicové vrstvy + ovládací panel ---
folium.TileLayer(
    'cartodbdark_matter',      # ID stylu
    name='Dark matter'         # název vrstvy v přepínači
).add_to(m)

folium.TileLayer(
    'cartodbpositron',
    name='Positron'
).add_to(m)

# Přepínač vrstev (panel vpravo nahoře)
folium.LayerControl().add_to(m)

# --- 7. Tři body ve Varšavě (souřadnice jsou orientační, ale reálné) ---
places = [
    {
        "location": [52.2318, 21.0059],
        "popup": "Krakowskie Przedmieście – historická ulice v centru",
        "tooltip": "Historické centrum"
    },
    {
        "location": [52.2319, 21.0067],
        "popup": "Royal Castle – Královský zámek",
        "tooltip": "Královský zámek"
    },
    {
        "location": [52.2299, 21.0122],
        "popup": "Warsaw Central Station – hlavní nádraží",
        "tooltip": "Centrální nádraží"
    },
]

for place in places:
    folium.Marker(
        location=place["location"],
        popup=folium.Popup(place["popup"], max_width=250),
        tooltip=place["tooltip"],
    ).add_to(m)

# --- 8. Zobrazení mapy v Colabu ---
fig
```

Vysvětlení krok za krokem

Figure:
Figure je „plátno“, do kterého přidáváš mapu. Umožňuje řídit rozměry výstupu (šířka/výška), aby mapa nebyla ani moc malá, ani moc velká.

Map:
folium.Map(location=..., zoom_start=..., tiles=None)

location – střed mapy (zeměpisná šířka, délka).

zoom_start – počáteční přiblížení (větší číslo = větší zoom).

tiles=None – neaktivujeme výchozí dlaždice, abychom mohli ručně přidat dvě vrstvy dle zadání.

TileLayer:
folium.TileLayer('cartodbdark_matter', name='Dark matter')

cartodbdark_matter a cartodbpositron jsou předdefinované styly (tmavý a světlý).

name je popisek, který se zobrazí v přepínači vrstev.

LayerControl:
Přidá panel v pravém horním rohu, kde si můžeš přepínat, jestli chceš tmavou nebo světlou mapu (nebo obě).

Marker + popup + tooltip:

Marker ukáže konkrétní bod.

popup se otevře po kliknutí (vhodné na delší text).

tooltip se zobrazí po najetí myší (krátký popisek).
Díky kombinaci obou má uživatel rychlou informaci (tooltip) i detailnější text (popup).

Krátká interpretace výsledku

Vidíš interaktivní mapu Varšavy, kde:

po najetí na značku se ukáže krátký tooltip,

po kliknutí se otevře detailní popup s popisem místa.

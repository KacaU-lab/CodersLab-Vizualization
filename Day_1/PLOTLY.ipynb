```python
# ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
# üß™ TEST Plotly instalace
# ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
import plotly.express as px
import pandas as pd

# Vytvo≈ô√≠me jednoduch√Ω DataFrame
df = pd.DataFrame({
    "Mƒõs√≠c": ["Leden", "√önor", "B≈ôezen", "Duben", "Kvƒõten"],
    "Prodeje": [120, 150, 170, 160, 190]
})

# Interaktivn√≠ sloupcov√Ω graf
fig = px.bar(df,
             x="Mƒõs√≠c",
             y="Prodeje",
             title="üìä Test Plotly: Prodeje za mƒõs√≠ce",
             color="Prodeje",
             text="Prodeje")

fig.update_traces(textposition='outside')
fig.update_layout(yaxis_title="Poƒçet prodan√Ωch kus≈Ø",
                  xaxis_title="Mƒõs√≠c",
                  title_x=0.5)

fig.show()
```
